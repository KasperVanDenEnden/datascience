# Numerical Corr Table

```python
AnimeNumerical = anime[['Episodes','Duration_Minutes','Score','Scored_Users','Ranked','Popularity','Members','Favorites']].dropna().corr()
AnimeNumerical.style.background_gradient(cmap='coolwarm', axis=None).set_precision(2)
```

Ik had inderdaad lage corrolaties verwacht als het gaat met betrekking tot episodes en duration_minutes. Dit komt omdat veel anime's dezelfde aantal afleveringen bevatten en afleveringen voor anime standaard rond de 20-25 minuten zijn. Daarom zal dit geen oorzaak zijn voor andere gegevens.
Verder is het wel logisch dat er een sterke corrolatie is tussen score en Ranked. Hoe hoger de score hoe hoger je geranked staat natuurlijk
